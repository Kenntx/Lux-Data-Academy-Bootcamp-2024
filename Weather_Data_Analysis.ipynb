```python
# Records where weather was clear

clear_weather = data[data['Weather'] == "Clear"]
print(clear_weather)
```

```python
# number of times the wind speed was exactly 4km/hr

wind_speed_4 = data[data['Wind Speed_km/h'] == 4]
print(wind_speed_4)
```

```python
# Check if there are any null values in the data set

data.isnull()
```

```python
# Rename the column Weather to Weather_Condition

data.rename(columns={'Weather' : 'Weather_Condition'}, inplace = True) 
```

```python
mean_vis = data['Visibility_km'].mean()

print(mean_vis)
```

```python
# Number of records where the wind speed is greater than 24km/h and visibility is equal to 25km

# Filter the data
speed_vis = data[data['Wind Speed_km/h'] > 24 & (data['Visibility_km'] == 25)]

# Count the records
record_count = speed_vis.shape[0]
print(record_count)
```

```python
# Mean value of each column for each weather condition.

mean_values = data.groupby('Weather').mean(numeric_only = True)
print(mean_values)
```

```python
# All instances where the weather is clear and the relative humidity is greater than 50 or visibility is above 40

filtered_data = data[(data['Weather'] == 'Clear') & ((data['Rel Hum_%']) > 50 | (data['Visibility_km'] > 40))]
print(filtered_data)
```

```python
# Number of weather conditions that include snow

snow_cond = data[data['Weather'].str.contains('Snow', case = False, na = False)]

unique_snow_cond = snow_cond['Weather'].nunique()
print(unique_snow_cond)
```

```python
# Rename the column Weather to Weather_Condition

data.rename(columns={'Weather' : 'Weather_Condition'}, inplace = True) 
```

```python
mean_vis = data['Visibility_km'].mean()

print(mean_vis)
```

```python
# Number of records where the wind speed is greater than 24km/h and visibility is equal to 25km

# Filter the data
speed_vis = data[data['Wind_Speed_kmh'] > 24 & (data['Visibility_km'] == 25)]

# Count the records
record_count = speed_vis.shape[0]
print(record_count)
```

```python
# Mean value of each column for each weather condition.

mean_values = data.groupby('Weather_Condition').mean(numeric_only = True)
print(mean_values)
```

```python
# All instances where the weather is clear and the relative humidity is greater than 50 or visibility is above 40

filtered_data = data[(data['Weather_Condition'] == 'Clear') & ((data['Rel_Hum_percent']) > 50 | (data['Visibility_km'] > 40))]
print(filtered_data)
```

```python
# Number of weather conditions that include snow

snow_cond = data[data['Weather_Condition'].str.contains('Snow', case = False, na = False)]

unique_snow_cond = snow_cond['Weather_Condition'].nunique()
print(unique_snow_cond)
```

```python
# Identify and describe any significant outliers in the data.
# For this task we will create boxplots for numerical features

numerical_columns = data.select_dtypes(include='number').columns

for column in numerical_columns:
    plt.figure(figsize = (8, 6))
    sns.boxplot(x = data[column])
    plt.title(f'Box Plot of {column}')
    plt.xlabel(column)
    plt.show
    
```

```python
# Conclusions
# Temperature: Extremes of -23.3Â°C and 33.0Â°C could be worth further investigation, especially if they correspond to unusual weather events.
# Dew Point: Values as low as -28.5Â°C and as high as 24.4Â°C may indicate extreme moisture conditions.
# Wind Speed: A maximum of 83 km/h suggests possible storm events or very windy conditions.
# Visibility: Minimum visibility of 0.2 km could indicate severe fog or other obstructive weather conditions.
```

```python
#  Create visualizations to show the distribution of key weather parameters (e.g., temperature, humidity, wind speed).

# Create histograms for temperature, humidity, and wind speed

plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)
sns.histplot(data=data, x='Temp_C')
plt.title('Temperature Distribution')

plt.subplot(2, 2, 2)
sns.histplot(data=data, x='Rel_Hum_percent')
plt.title('Humidity Distribution')

plt.subplot(2, 2, 3)
sns.histplot(data=data, x='Wind_Speed_kmh')
plt.title('Wind Speed Distribution')

plt.tight_layout()
plt.show()
```

```python
data['Date_Time'] = pd.to_datetime(data['Date_Time'])
```

```python
# Create time series graphs to visulaize trends over time. Highlight any notable patterns or seasonal variations.

# Plotting time series for Temperature, Humidity, and Wind Speed

```

```python
# Temperature over time
#plt.subplot(3, 1, 1)
sns.lineplot(data=data, x='Date_Time', y='Temp_C', color='skyblue')
plt.title('Temperature over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (Â°C)')

plt.show()
```

```python
# Relative Humidity over time
#plt.subplot(3, 1, 2)
sns.lineplot(data=data, x='Date_Time', y='Rel_Hum_percent', color='lightgreen')
plt.title('Relative Humidity over Time')
plt.xlabel('Date')
plt.ylabel('Relative Humidity (%)')

plt.show()
```

```python
# Wind Speed over time
# plt.subplot(3, 1, 3)
sns.lineplot(data=data, x='Date_Time', y='Wind_Speed_kmh', color='salmon')
plt.title('Wind Speed over Time')
plt.xlabel('Date')
plt.ylabel('Wind Speed (km/h)')

```

```python
# Drop non-numeric columns
numerical_data = data.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix for numerical variables
corr_matrix = numerical_data.corr()

# Plot the heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Weather Parameters')
plt.show()
```

```python
# Extract month and year for seasonal analysis
data['Month'] = data['Date_Time'].dt.month
data['Year'] = data['Date_Time'].dt.year

# Group by month and calculate the mean temperature and humidity
monthly_means = data.groupby('Month')[['Temp_C', 'Rel_Hum_percent']].mean()

# Plot seasonal trends for Temperature and Humidity
plt.figure(figsize=(12, 6))
monthly_means.plot(kind='line', marker='o', ax=plt.gca())
plt.title('Monthly Trends in Temperature and Humidity')
plt.xlabel('Month')
plt.ylabel('Mean Value')
plt.legend(['Temperature (Â°C)', 'Relative Humidity (%)'])
plt.grid(True)
plt.show()
```
